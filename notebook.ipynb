### Assessing Data

```python
#Cek Type data
df.info()
```

```python
#identifikasi missing value
df.isna().sum()
```

```python
#Identifikasi Duplicate data
print("Jumlah Duplikasi : " ,df.duplicated().sum())
```

```python
df.describe()
```

```python
#Identifikasi keakuratan Data Value dimana jika casual + registered != cnt maka data tidak akurat
mean_cnt = df['cnt'].mean()
mean_registered = df['registered'].mean()
mean_casual = df['casual'].mean()

print(f"Rata-rata 'cnt': {mean_cnt:.6f}")
print(f"Rata-rata 'registered': {mean_registered:.6f}")
print(f"Rata-rata 'casual': {mean_casual:.6f}")

# Cek konsistensi antara casual, registered, dan cnt
inconsistent_data = df[df['casual'] + df['registered'] != df['cnt']]

# Menampilkan hasil dengan keterangan
if inconsistent_data.empty:
    print("Data yang tidak konsisten: Tidak ada. Semua baris konsisten.")
else:
    print("Data yang tidak konsisten:")
    print(inconsistent_data)

# Menghitung persentase rata-rata
total_average = mean_registered + mean_casual
percentage_difference = ((mean_cnt - total_average) / mean_cnt) * 100

print(f"\nSelisih rata-rata antara 'cnt' dan jumlah rata-rata 'registered' + 'casual': {percentage_difference:.2f}%")
```

```python
#identifikasi Outlier
column = 'cnt'
q25, q75 = np.percentile(df[column], 25), np.percentile(df[column], 75)
iqr = q75 - q25

cut_off = iqr * 1.5
minimum, maximum = q25 - cut_off, q75 + cut_off

outliers = df[(df[column] < minimum) | (df[column] > maximum)]
print("Outliers berdasarkan metode IQR:")
print(outliers)
```

**Insight:**
- Dari hasil proses Assessing ada type data yang harus diubah yaitu dteday,karena Kolom dteday sebaiknya diubah menjadi tipe datetime agar lebih mudah dalam analisis berbasis waktu.
