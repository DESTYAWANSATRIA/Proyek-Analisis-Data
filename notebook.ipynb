```python
#Identifikasi keakuratan Data Value dimana jika casual + registered != cnt maka data tidak akurat
mean_cnt = hour_df['cnt'].mean()
mean_registered = hour_df['registered'].mean()
mean_casual = hour_df['casual'].mean()

print(f"Rata-rata 'cnt': {mean_cnt:.6f}")
print(f"Rata-rata 'registered': {mean_registered:.6f}")
print(f"Rata-rata 'casual': {mean_casual:.6f}")

# Cek konsistensi antara casual, registered, dan cnt
inconsistent_data = hour_df[hour_df['casual'] + hour_df['registered'] != hour_df['cnt']]

# Menampilkan hasil dengan keterangan
if inconsistent_data.empty:
    print("Data yang tidak konsisten: Tidak ada. Semua baris konsisten.")
else:
    print("Data yang tidak konsisten:")
    print(inconsistent_data)

# Menghitung persentase rata-rata
total_average = mean_registered + mean_casual
percentage_difference = ((mean_cnt - total_average) / mean_cnt) * 100
```

## Exploratory Data Analysis (EDA)

### Explore 
- Rata rata jumlah pengguna berdasarkan temperatur.

```python
#Menampilkan rata rata jumlah peminjam berdasarkan temperatur pada hour.csv
avg_hourly_borrowers_by_temp = hour_df.groupby('temp')['cnt'].mean().reset_index()
print(avg_hourly_borrowers_by_temp)
```

```python
#Analisis jumlah peminjam tiap bulan ke-1 hingga bulan ke-12 pada tahun 2011 dan 2012
hour_df['dteday'] = hour_df['dteday'].astype(str)
hour_df['year'] = hour_df['dteday'].str[:4] 
hour_df['month'] = hour_df['dteday'].str[5:7] 
monthly_borrowers = hour_df.groupby(['year', 'month'])['cnt'].sum().reset_index()
print(monthly_borrowers)
```

```python
#Menampilkan jumlah peminjam berdasarkan season
avg_hourly_borrowers_by_season = hour_df.groupby('season')['cnt'].sum().reset_index()
print(avg_hourly_borrowers_by_season)
```

```python
# Menghitung total peminjam casual dan registered 
total_casual = hour_df['casual'].sum()
print("Total Pengguna Casual : ",total_casual)

total_registered = hour_df['registered'].sum()
print("Total Pengguna registered : ",total_registered)
```

```python
# Menghitung rata-rata jumlah pengguna berdasarkan kondisi cuaca (weathersit)
weather_grouped = hour_df.groupby('weathersit')['cnt'].mean().reset_index()
print("Rata-rata Jumlah Pengguna Berdasarkan Kondisi Cuaca:")
print(weather_grouped)
```

```python
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
hour_df['year_month'] = hour_df['dteday'].dt.to_period('M')

# Menghitung jumlah pengguna terdaftar dan kasual per bulan
monthly_grouped = hour_df.groupby('year_month').agg({'registered': 'sum', 'casual': 'sum'}).reset_index()
print("\nJumlah Pengguna Terdaftar dan Kasual per Bulan:")
print(monthly_grouped)
```

```python
# Menghitung persentase perubahan pengguna terdaftar dan kasual per bulan
monthly_grouped['registered_change'] = monthly_grouped['registered'].pct_change() * 100
monthly_grouped['casual_change'] = monthly_grouped['casual'].pct_change() * 100

print("\nPersentase Perubahan Pengguna Terdaftar dan Kasual per Bulan:")
print(monthly_grouped[['year_month', 'registered_change', 'casual_change']])
```

```python
# Menghitung rata-rata jumlah pengguna berdasarkan jenis hari
avg_users = hour_df.groupby('workingday')['cnt'].mean().reset_index()
print("\nRata-rata Jumlah Pengguna Berdasarkan Hari Kerja/Hari Libur:")
print(avg_users)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana kita bisa memprediksi tren penggunaan sepeda di kondisi tertentu berdasarkan data historis?

```python
plt.figure(figsize=(10, 6))
plt.scatter(avg_hourly_borrowers_by_temp['temp'], avg_hourly_borrowers_by_temp['cnt'], color='blue', alpha=0.7)

# Menambahkan judul dan label
plt.title('Average Bike Rentals based on Temperature (Hour)', fontsize=14)
plt.xlabel('Temperature (temp)', fontsize=12)
plt.ylabel('Average Rentals (cnt)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_borrowers, x='month', y='cnt', hue='year', palette='viridis')
plt.title('Jumlah Peminjam Tiap Bulan pada Tahun 2011 dan 2012')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peminjam')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])
plt.legend(title='Tahun')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Membuat figure dan plot bar untuk analisis jumlah peminjam tiap season
plt.figure(figsize=(6, 5))
plt.bar(avg_hourly_borrowers_by_season['season'], avg_hourly_borrowers_by_season['cnt'], 
        color=["#D3D3D3", "#D3D3D3", "#87CEFA", "#D3D3D3"])

plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.title('Jumlah Peminjam Berdasarkan Season', fontsize=14)
plt.xticks([1, 2, 3, 4], ['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

### Pertanyaan 2: Strategi apa yang digunakan untuk meningkatkan pendapatan, kenyamanan dan pengalaman pengguna?

```python
# Menghitung rata-rata jumlah pengguna berdasarkan kondisi cuaca (weathersit)
weather_grouped = hour_df.groupby('weathersit')['cnt'].mean().reset_index()

# Membuat bar plot
plt.figure(figsize=(10, 6))
plt.bar(weather_grouped['weathersit'].astype(str), weather_grouped['cnt'], color='#87CEFA')
plt.title('Rata-rata Jumlah Pengguna Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca (1: Cerah, 2: Mendung, 3: Hujan)')
plt.ylabel('Rata-rata Jumlah Pengguna')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
# Mengambil bulan dan tahun dari tanggal
hour_df['year_month'] = hour_df['dteday'].dt.to_period('M')

# Menghitung jumlah pengguna terdaftar dan kasual per bulan
monthly_grouped = hour_df.groupby('year_month').agg({'registered': 'sum', 'casual': 'sum'}).reset_index()

# Membuat stacked bar chart per bulan
plt.figure(figsize=(12, 6))
plt.bar(monthly_grouped['year_month'].astype(str), monthly_grouped['registered'], label='Pengguna Terdaftar', color='#87CEFA')
plt.bar(monthly_grouped['year_month'].astype(str), monthly_grouped['casual'], 
        bottom=monthly_grouped['registered'], label='Pengguna Kasual', color='#D3D3D3')
plt.title('Perbandingan Pengguna Kasual dan Terdaftar per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pengguna')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_grouped['registered_change'] = monthly_grouped['registered'].pct_change() * 100
monthly_grouped['casual_change'] = monthly_grouped['casual'].pct_change() * 100
```

```python
plt.figure(figsize=(12, 6))
plt.bar(monthly_grouped['year_month'].astype(str), monthly_grouped['registered_change'], label='Perubahan Pengguna Terdaftar (%)', color='#87CEFA')
plt.bar(monthly_grouped['year_month'].astype(str), monthly_grouped['casual_change'], label='Perubahan Pengguna Kasual (%)', color='#D3D3D3', bottom=monthly_grouped['registered_change'])
plt.title('Persentase Perubahan Pengguna Terdaftar dan Kasual per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Persentase Perubahan (%)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Menghitung rata-rata jumlah pengguna berdasarkan jenis hari
avg_users = hour_df.groupby('workingday')['cnt'].mean().reset_index()

# Membuat bar plot
plt.figure(figsize=(8, 5))
plt.bar(avg_users['workingday'].astype(str), avg_users['cnt'], color=['#87CEFA', '#D3D3D3'])
plt.title('Rata-rata Jumlah Pengguna Berdasarkan Hari Kerja/Hari Libur')
plt.xlabel('Hari Kerja (0: Libur, 1: Kerja)')
plt.ylabel('Rata-rata Jumlah Pengguna')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- Dengan melihat rata-rata jumlah pengguna berdasarkan kondisi cuaca, Anda dapat memahami bagaimana cuaca memengaruhi jumlah pengguna. Ini dapat membantu dalam perencanaan, seperti meningkatkan layanan pada hari cuaca cerah yang cenderung lebih banyak pengguna. Strateginya kita bisa mempertimbangkan untuk meluncurkan promosi atau kegiatan khusus pada hari-hari dengan cuaca baik untuk menarik lebih banyak pengguna.

Kita dapat melihat bulan mana yang memiliki lebih banyak pengguna terdaftar atau kasual. Jika ada perubahan besar, Kita bisa menggali lebih dalam untuk mencari penyebabnya (misalnya, promosi yang berhasil atau faktor musiman).
