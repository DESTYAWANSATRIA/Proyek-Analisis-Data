### Explore ...

```python
#Menampilkan rata rata jumlah peminjam berdasarkan temperatur pada hour.csv
avg_hourly_borrowers_by_temp = hour_df.groupby('temp')['cnt'].mean().reset_index()
print(avg_hourly_borrowers_by_temp)
```

```python
#Analisis jumlah peminjam tiap bulan ke-1 hingga bulan ke-12 pada tahun 2011 dan 2012
hour_df['dteday'] = hour_df['dteday'].astype(str)
hour_df['year'] = hour_df['dteday'].str[:4] 
hour_df['month'] = hour_df['dteday'].str[5:7] 
monthly_borrowers = hour_df.groupby(['year', 'month'])['cnt'].sum().reset_index()
print(monthly_borrowers)
```

```python
#Menampilkan jumlah peminjam berdasarkan season
avg_hourly_borrowers_by_season = hour_df.groupby('season')['cnt'].sum().reset_index()
print(avg_hourly_borrowers_by_season)
```

```python
# Menghitung total peminjam casual dan registered 
total_casual = hour_df['casual'].sum()
print("Total Pengguna Casual : ",total_casual)

total_registered = hour_df['registered'].sum()
print("Total Pengguna registered : ",total_registered)
```
